```python
import random

suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', "Five", 'Six', 'Seven', "Eight", 'Nine', 'Ten', 'Jack', 'Queen',
         'King', 'Ace')
values = {'Two': 2, 'Three': 3, 'Four': 4, 
         "Five": 5, 'Six': 6, 'Seven':7, "Eight": 8,
         'Nine': 9, 'Ten': 10, 'Jack': 11, 'Queen': 12,
         'King': 13, 'Ace': 14}

playing = True
```

```python
#create a card class that would return rank and suit of a card 
class Card:
    def __init__(self, suit, rank):
        self.suit = suit
        self.rank = rank
    
    def __str__(self):
        return self.rank + ' of ' + self.suit
```

```python
class Deck:
    def __init__(self):
        self.all_cards = []
        
        for suit in suits:
            for rank in ranks:
                created_card= Card(suit, rank)
                #create a card object
                self.all_cards.append(created_card)
  #create a shuffle method   
    def shuffle_deck(self):
        
        random.shuffle(self.all_cards)          
    #pop() a card from the deck to add into the users deck 
    def deal_one(self):
        return self.all_cards.pop()
```

```python
#create a hand class that woudl also be able to calculate the value of cards and adjust the value of aces when needed 

class Hand:
    def __init__(self):
        self.cards = []  # start with an empty list as we did in the Deck class
        self.value = 0   # start with zero value
        self.aces = 0    # add an attribute to keep track of aces
    
    def add_card(self, card):
        self.cards.append(card) # card passed in is gonna be from the deck object, deck.deal()
        self.value += values[card.rank]
       #check for the aces  
        if card.rank == "Ace":
            self.aces += 1
        
        #if the value is >21, then it subtracts 10 from the value, equating the ace to 1. 
        #since the ace is no longer 11, I also remove it from the count. 
    def adjust_for_ace(self):
        while self.value > 21 and self.aces:
            self.value -= 10
            self.aces -= 1
```

```python
#either substract the best if they lose or add if they win 
class Chips:
    def __init__(self):
        self.total = 100
        self.bet = 0
        
    def win_bet(self):
        self.total += self.bet
    
    def lose_bet(self):
        self.total -= self.bet
```

### functions for game play 

```python
#func that takes a user's bet and adjusts it as an actual bet 
def take_bet(chips):
    while True:
        try:
            chips.bet = int(input("How many chips ya wanna bet? "))
            
        except:
            print("Please, put an INTEGER!")
            
        else:
            if chips.bet > chips.total:
                print("Sorz, you do not have enough chips, meh. You have {}".format(chips.total))
            
            else:
                break
        

```

```python
#func for taking hits taking hand and deck as args
def hit(deck,hand):
    
    single_card = deck.deal_one()
    hand.add_card(single_card)
    hand.adjust_for_ace()
```

```python
#If the Player Hits, employ the hit() function above. 
#If the Player Stands, set the playing variable to False 
def hit_or_stand(deck,hand):
    global playing  # to control an upcoming while loop
    
    while True:
        x = input("Hit or Stand? Enter h or s")
        if x[0].lower() == 'h':
            hit(deck,hand)
            
        elif x[0].lower() == 's':
            print("Player stands, dealer's turn")
            playing = False
            
        else:
            print("Please, enter h or s only")
            continue
        
        break
```
